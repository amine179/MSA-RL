```python
# save all_rewards to csv file
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
all_rewards = pd.read_csv('all_rewards.txt', sep='\t', header=None)
all_rewards.head()
```

```python
def moving_average(values, window):
    weights = np.repeat(1.0, window) / window
    ma = np.convolve(values, weights, 'valid')
    return ma
```

```python
def plot_rs(sqn):
    x = np.array(all_rs[sqn])
    y = moving_average(x, 600)
    plt.title(" rewards for sequences numebr {}".format(sqn))
    plt.plot(y)
    plt.show()
```

```python
for n in range(0, 25):
    plot_rs(n)
#     break #break ila bghiti ghir plot lowl
```

```python
from Bio.Align import MultipleSeqAlignment

from Bio.Seq import Seq
from Bio.SeqRecord import SeqRecord

# Define the alignment
alignment = MultipleSeqAlignment([
    SeqRecord(Seq("AKVVA"), id="seq1"),
    SeqRecord(Seq("CVK--"), id="seq2"),
    SeqRecord(Seq("AK---"), id="seq3")
])

# Define the scoring matrix
from Bio.SubsMat import MatrixInfo
matrix = MatrixInfo.blosum62

# Calculate the sum of pairs score for each column in the alignment
scores = []
for i in range(alignment.get_alignment_length()):
    column = alignment[:, i]
    pairs = [(column[j], column[k]) for j in range(len(column)) for k in range(j+1, len(column))]
    matches = sum(matrix.get(pair, matrix.get(tuple(reversed(pair)), -4)) for pair in pairs)
    score = matches / len(pairs)
    scores.append(score)

# Calculate the average sum of pairs score for the whole alignment
average_score = sum(scores) / len(scores)
print("Average sum of pairs score:", average_score)
```
